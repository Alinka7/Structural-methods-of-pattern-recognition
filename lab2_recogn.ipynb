```python
import re
import string
import nltk
import itertools 
from collections import Counter
import math
import random
import numpy as np
```

```python
def prepr1(data):
    op_string = re.sub(f"[{re.escape(string.punctuation)}]", "", data)
    op_string = re.sub(' +', ' ', op_string)
    op_string = op_string.replace('\n', '').lower()
    listWord = op_string.split(' ')
    return listWord
    
def prepr2(data):    
    listWord = [' ' +i  if i != data[0] else i for i in data]
    listWord = [i+' '  if i !=  listWord[-1] else i for i in listWord]
    r = []
    rUpd = []
    for i in listWord:
        r.append(list(nltk.bigrams(i)))
    for i in r:
        rUpd.append([''.join(list(j)) for j in i])    
    rUpd = [x for x in rUpd if x]
    totalResult = list(itertools.chain(*rUpd))
    return totalResult
```

```python
def extractPhrase(testData):
    r = testData[random.randint(0, int(len(testData)/2)):random.randint(int(len(testData)/2), int(len(testData)))]
    while (len(r) > 5):
        r = r[0:int(len(r)/2)]
        if len(r) < 8 and len(r) > 3:
#             print(r)
            return r
```

```python
def creatResDict(totalDict, listBigramsPhrase):
    test = {}
    for elem in listBigramsPhrase:
        for k, v in totalDict.items():
            if elem == k:
                test[elem]= v
    return test
```

```python
def calcCoef(si, pi):
    s_avg = sum(si) / len(si)
    p_avg = sum(pi) / len(pi)
    r = (s_avg + p_avg)/2
    return r
```

```python
import numpy as np
from PIL import Image
import cv2

import warnings
warnings.filterwarnings('ignore')
```

```python
image = cv2.imread("dog_1.jpg", cv2.IMREAD_COLOR)

height, width, channels = image.shape
    
gradient_x = np.zeros((height, width, channels))
gradient_y = np.zeros((height, width, channels))
  
```

```python
image = image.astype('float')
```

```python
divergence = np.zeros((height, width, channels))
divergence.shape
```

```python
divergence = np.zeros((height, width, channels))
for c in range(channels):
    for i in range(1, height-1):
        for j in range(1, width-1):
            divergence[i, j, c] = (image[i-1, j, c] + image[i+1, j, c] + image[i, j-1, c] + image[i, j+1, c] - 4*image[i, j, c])                
```

```python
cv2.imshow('divergence',divergence)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('divergence.jpg', divergence)
```

```python
m = height
n = width
boundary = np.zeros((height, width, channels))
r_boundary = np.zeros((gradient_x.shape[0], gradient_x.shape[1]))
r_boundary[:, 0] = image[:, 0, 0]
r_boundary[:, m-1] = image[:, m-1, 0]
r_boundary[0, :] = image[0, :, 0] 
r_boundary[n-1, :] =  image[n-1, :, 0]

g_boundary  = np.zeros((gradient_x.shape[0], gradient_x.shape[1]))
g_boundary [:, 0] =image[:, 0, 1]
g_boundary [:, m-1] = image[:, m-1, 1]
g_boundary [0, :] =  image[0, :, 1] 
g_boundary [n-1, :] = image[n-1, :, 1]

b_boundary  = np.zeros((gradient_x.shape[0], gradient_x.shape[1]))
b_boundary [:, 0] =  image[:, 0, 2]
b_boundary [:, m-1] =  image[:, m-1, 2]
b_boundary [0, :] =  image[0, :, 2] 
b_boundary [n-1, :] = image[n-1, :, 2]
```

```python
bounds = np.stack([r_boundary, g_boundary, b_boundary], axis=-1)
```

```python
restored_image = bounds
result = np.zeros_like(image)
for c in range(3):
    channel_divergence = divergence[:, :, c]
    max_iterations = 5000
    # 
    tolerance = 1e-5
    
    previous_image = np.copy(restored_image[:, :, c])
    for _ in range(max_iterations):
        maxDif = 0
        for i in range(1, image.shape[0] - 1):
            for j in range(1, image.shape[1] - 1):    
                temp = (previous_image[i + 1, j] + previous_image[i - 1, j] + previous_image[i, j + 1] + previous_image[i, j - 1])/4 - channel_divergence[i, j]/4      
                difference = np.abs(previous_image[i, j] - temp)
                previous_image[i, j] = temp

                if difference > maxDif:
                    maxDif = difference
        print(maxDif) 
        print('iter', _)
        if maxDif < tolerance:
            break
            
    result[:, :, c] = previous_image
    
    
result = result.astype('uint8')
cv2.imshow('Restored Image', result)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('task1.jpg', result)
```
